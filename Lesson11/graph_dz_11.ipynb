```python
import scipy.optimize
import sympy as sp
from sympy import cos, sin
x = sp.Symbol('x')
expr =-12*x**4*sin(cos(x))-18*x**3+5*x**2 + 10*x - 30
```

```python
print('Исходная функция')
print(expr)
expr_diff = sp.diff(expr)
sp.plot(expr, (x, -8, 8))
print('Производная функция')
print(expr_diff)
sp.plot(expr_diff, (x, -8, 8))
```
